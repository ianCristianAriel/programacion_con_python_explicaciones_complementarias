````python
class Producto: # Usaremos este nombre de ahora en adelante
    """Representa un producto del supermercado con datos y acciones."""
    def __init__(self, codigo_barra, nombre, precio, stock_inicial):
        # Los atributos son los mismos que antes
        self.codigo_barra = codigo_barra
        self.nombre = nombre
        self.precio = precio
        self.stock = stock_inicial
        # print(f"-> Ejecutando __init__ para Producto: Creando '{nombre}'") # Opcional

    # --- NUEVOS MÉTODOS ---
    def actualizar_stock(self, cantidad):
        """Actualiza el stock del producto. Positivo para ingreso, negativo para venta/salida."""
        if self.stock + cantidad >= 0:
            self.stock += cantidad
            print(f"Stock de '{self.nombre}' actualizado. Nuevo stock: {self.stock}")
        else:
            print(f"Error: No hay suficiente stock de '{self.nombre}' para quitar {abs(cantidad)}. Stock actual: {self.stock}")

    def aplicar_descuento(self, porcentaje_descuento):
        """Aplica un descuento al precio del producto. El porcentaje debe estar entre 0 y 100."""
        if 0 <= porcentaje_descuento <= 100:
            descuento = self.precio * (porcentaje_descuento / 100)
            self.precio -= descuento
            print(f"Descuento del {porcentaje_descuento}% aplicado a '{self.nombre}'. Nuevo precio: ${self.precio:.2f}")
        else:
            print("Error: El porcentaje de descuento debe estar entre 0 y 100.")

    def obtener_info_display(self):
        """Devuelve un string con la información formateada del producto para mostrar."""
        # Accede a los atributos del propio objeto usando self
        return f"[{self.codigo_barra}] {self.nombre} - Precio: ${self.precio:.2f} - Stock: {self.stock}"
````

````python
#  Modificar la clase Producto para agregar el método calcular_valor_inventario
class Producto:
    """Representa un producto del supermercado con datos y acciones."""
    def __init__(self, codigo_barra, nombre, precio, stock_inicial):
        self.codigo_barra = codigo_barra
        self.nombre = nombre
        self.precio = precio
        self.stock = stock_inicial

    def actualizar_stock(self, cantidad):
        """Actualiza el stock del producto. Positivo para ingreso, negativo para venta/salida."""
        if self.stock + cantidad >= 0:
            self.stock += cantidad
            print(f"Stock de '{self.nombre}' actualizado. Nuevo stock: {self.stock}")
        else:
            print(f"Error: No hay suficiente stock de '{self.nombre}' para quitar {abs(cantidad)}. Stock actual: {self.stock}")

    def aplicar_descuento(self, porcentaje_descuento):
        """Aplica un descuento al precio del producto. El porcentaje debe estar entre 0 y 100."""
        if 0 <= porcentaje_descuento <= 100:
            descuento = self.precio * (porcentaje_descuento / 100)
            self.precio -= descuento
            print(f"Descuento del {porcentaje_descuento}% aplicado a '{self.nombre}'. Nuevo precio: ${self.precio:.2f}")
        else:
            print("Error: El porcentaje de descuento debe estar entre 0 y 100.")

    def obtener_info_display(self):
        """Devuelve un string con la información formateada del producto para mostrar."""
        return f"[{self.codigo_barra}] {self.nombre} - Precio: ${self.precio:.2f} - Stock: {self.stock}"

    def calcular_valor_inventario(self):
        """Calcula el valor total del stock actual (precio * stock)."""
        return self.precio * self.stock
````

````python
# Crear una instancia de Producto
prod = Producto(codigo_barra="003", nombre="Leche", precio=150.0, stock_inicial=50)
````

````python
# Calcular valor del inventario y mostrarlo
valor_inventario = prod.calcular_valor_inventario()
print(f"Valor total del stock de '{prod.nombre}': ${valor_inventario:.2f}")
````

````python
# Actualizar stock
prod.actualizar_stock(-10)  # Se venden 10 unidades
````

````python
# Volver a calcular valor del inventario
valor_inventario = prod.calcular_valor_inventario()
print(f"Valor total del stock de '{prod.nombre}' después de la venta: ${valor_inventario:.2f}")
````

#### Solución ejercicio 3:

````python
# Clase CarritoDeCompras
class CarritoDeCompras:
    """Representa un carrito de compras con productos y cantidades."""

    def __init__(self):
        self.items = []  # Lista vacía de items

    def agregar_item(self, producto_obj, cantidad=1):
        """Agrega un producto al carrito con la cantidad indicada."""
        if not isinstance(producto_obj, Producto):
            print("Error: El objeto agregado debe ser una instancia de Producto.")
            return
        if cantidad <= 0:
            print("Error: La cantidad debe ser positiva.")
            return
        if producto_obj.stock < cantidad:
            print(f"Error: No hay suficiente stock de '{producto_obj.nombre}' para agregar {cantidad} unidades. Stock actual: {producto_obj.stock}")
            return

        # Todo OK, agregar al carrito y descontar stock
        self.items.append({'producto': producto_obj, 'cantidad': cantidad})
        producto_obj.actualizar_stock(-cantidad)
        print(f"Agregado {cantidad} x '{producto_obj.nombre}' al carrito.")

    def calcular_total(self):
        """Calcula el total del carrito."""
        return sum(item['producto'].precio * item['cantidad'] for item in self.items)

    def mostrar_contenido(self):
        """Muestra los items del carrito y su subtotal."""
        if not self.items:
            print("El carrito está vacío.")
            return
        print("Contenido del carrito:")
        for item in self.items:
            prod = item['producto']
            cant = item['cantidad']
            subtotal = prod.precio * cant
            print(f"{prod.nombre} x {cant} = ${subtotal:.2f}")
        print(f"Total del carrito: ${self.calcular_total():.2f}")
````

````python
# Crear algunos productos
prod1 = Producto("101", "Leche", 150.0, 50)
prod2 = Producto("102", "Pan", 80.0, 10)
prod3 = Producto("103", "Huevos", 200.0, 0)  # Sin stock para probar error
````

````python
# Crear carrito
carrito = CarritoDeCompras()
````

````python
# Agregar items
carrito.agregar_item(prod1, 5)   # OK
carrito.agregar_item(prod2, 15)  # Error: no hay stock suficiente
carrito.agregar_item(prod3, 1)   # Error: sin stock
carrito.agregar_item(prod2, 5)   # OK
````

````python
# Mostrar contenido y total
carrito.mostrar_contenido()
````

````python
# Verificar stock final de productos
print(f"Stock final de '{prod1.nombre}': {prod1.stock}")
print(f"Stock final de '{prod2.nombre}': {prod2.stock}")
print(f"Stock final de '{prod3.nombre}': {prod3.stock}")
````
